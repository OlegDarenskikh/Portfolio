## Приоритизация гипотез

 В нашем распоряжении файл с 9 гипотезами, для которых оценены значения `Reach`, `Impact`, `Confidence` и `Efforts`. Применим к этим оценкам фреймворки `ICE` и `RICE`.

Составим профили пользователей. Определим минимальную и максимальную даты привлечения пользователей.

```python
# Подсчитаем важность гипотез по методу `ICE`

hypothesis['ICE'] = hypothesis['impact'] * hypothesis['confidence'] / hypothesis['efforts']

hypothesis[['hypothesis', 'ICE']].sort_values(by='ICE', ascending=False).head(5)
```

```python
# Построим гистограмму оценки гипотез по методу ICE

ax = (hypothesis[['hypothesis', 'ICE']]
      .set_index('hypothesis')
      .sort_values('ICE')
      .plot(kind='barh', alpha=0.4)
     )

ax.grid(True)
ax.set_xlabel('Приоритет')
ax.set_ylabel(' ')
ax.set_title('Приоритезация гипотез по методу ICE', loc='left')

plt.show()
```

Мы проранжировали гипотезы и уже становится понятно, на какие стоит оратить внимание. Добавим к расчету параметр Reach (Охват) и посчитаем метрику RICE.

```python
# Подсчитаем важность гипотез по методу `RICE`

hypothesis['RICE'] = hypothesis.reach * hypothesis.impact * hypothesis.confidence / hypothesis.efforts

display(hypothesis[['hypothesis', 'RICE']].sort_values(by='RICE', ascending=False))
```

```python
# Построим гистограмму оценки гипотез по методу RICE

ax = (hypothesis[['hypothesis', 'RICE']]
      .set_index('hypothesis')
      .sort_values('RICE')
      .plot(kind='barh', alpha=0.4)
     )
ax.grid(True)
ax.set_xlabel('Приоритет')
ax.set_ylabel(' ')
ax.set_title('Приоритезация гипотез по методу RICE', loc='left')

plt.show()
```

Видно, что изменилась приоретизация гипотез, разница заключается, в том что одна методика учитывает количество пользователей, а другая нет.

```python
# Построим общую таблицу

hypothesis[['hypothesis','ICE', 'RICE']].sort_values(by='RICE', ascending=False).head(5)
```

```python
# Визуализируем полученные данные

ax = (hypothesis[['hypothesis','ICE', 'RICE']]
      .set_index('hypothesis')
      .sort_values('RICE')
      .plot(kind='barh', alpha=0.4)
     )
ax.grid(True)
ax.set_xlabel('Приоритет')
ax.set_ylabel(' ')
ax.set_title('Приоритезация гипотез по методам ICE и RICE', loc='left')

plt.show()
```

## Анализ A/B-теста

```python
# Найдем пользователей попавших в обе группы

double_visitors = orders.groupby(
    ['visitor_id']).agg(
    {
        'group' : 'nunique', 
        'transaction_id' : 'count', 
        'revenue' : 'sum'
    }).reset_index()
```

```python
print(f'Количество пользователей в 2-х группах: {double_visitors[double_visitors.group != 1]["visitor_id"].count()}')
print(f"Сумма заказов этих пользователей - {double_visitors[double_visitors.group != 1]['revenue'].sum()}")
print(f"Количество заказов этих пользователей - {double_visitors[double_visitors.group != 1]['transaction_id'].sum()}")
```

Теперь видно, что 58 человек попали в обе группы, они сделали 181 заказ на сумму более 1,5 млн., мы не можем сказать, что именно повлияло на их выбор, поэтому просто исключим их из исследования.

```python
# Занесем всех задвоившихся пользователей в список

double_list = list(double_visitors[double_visitors.group != 1]['visitor_id'])

# Перезапишем данные таблицы orders

orders = orders.query('visitor_id not in @double_list')
```

### Кумулятивные метрики

Чтобы построить графики по кумулятивным (накапливаемым) данным, необходимо собрать соответствующий агрегированный датафрейм, содержащий информацию о дате, группе (А или В), кумулятивных количестве заказов и пользователей их оформивших, среднем чеке и, наконец, о кумулятивном количестве посетителей магазина.

```python
# Создадим массив уникальных пар значений дат и групп теста
dates_grouped = orders[['date', 'group']].drop_duplicates()
```

```python
# Получим агрегированные кумулятивные данные о заказах по дням 

orders_aggregated = dates_grouped.apply(
    lambda x: orders[
        np.logical_and(
            orders['date'] <= x['date'], orders['group'] == x['group']
        )
    ].agg(
        {
            'date': 'max',
            'group': 'max',
            'transaction_id': 'nunique',
            'visitor_id': 'nunique',
            'revenue': 'sum',
        }
    ),
    axis=1,
).sort_values(by=['date', 'group'])

orders_aggregated.head()
```

```python
# Получим агрегированные данные о посетителях по дням

visitors_aggregated = dates_grouped.apply(
    lambda x: visitors[
        np.logical_and(
            visitors['date'] <= x['date'], visitors['group'] == x['group']
        )
    ].agg(
        {
            'date': 'max', 
            'group': 'max', 
            'visitors': 'sum'
        }
    ),
    axis=1,
).sort_values(by=['date', 'group'])

visitors_aggregated.head()
```

```python
# Обединим таблицы в одну и переименуем столбцы

cumulative_data = orders_aggregated.merge(
    visitors_aggregated, 
    left_on=['date', 'group'], 
    right_on=['date', 'group']
)
cumulative_data.columns = [
    'date',
    'group',
    'orders',
    'buyers',
    'revenue',
    'visitors',
]
```

```python
# Добавим расчетную колонку с конверсией

cumulative_data['conversion'] = cumulative_data['orders'] / cumulative_data['visitors']

cumulative_data.head()
```

Полученную таблицу мы будем использовать и далее.

### Кумулятивная выручка

```python
# Объявим датафрейм с датой, выручкой и количеством заказов для группы А

cumulative_revenue_a = cumulative_data[cumulative_data['group']=='A'][['date','revenue', 'orders']]

# Объявим датафрейм с датой, выручкой и количеством заказов для группы В

cumulative_revenue_b = cumulative_data[cumulative_data['group']=='B'][['date','revenue', 'orders']]

plt.plot(cumulative_revenue_a['date'], cumulative_revenue_a['revenue'], label='A')
plt.plot(cumulative_revenue_b['date'], cumulative_revenue_b['revenue'], label='B')
plt.title('График кумулятивной выручки по группам А и В')
plt.ylabel("Выручка, млн.")
plt.xticks(rotation=20)
plt.legend()
plt.show()
```

Выручка в обеих группах увеличивается на протяжении всего теста. Однако в группе B есть резкий скачок, который может свидетельствовать об увеличении числа заказов или о появлении неожидано дорогих заказов.

### Кумулятивный средний чек

Построим графики среднего чека по группам — разделим кумулятивную выручку на кумулятивное число заказов.

```python
# Рассчитаем средний чек для группы А

plt.plot(cumulative_revenue_a['date'], 
         cumulative_revenue_a['revenue'] / cumulative_revenue_a['orders'],
         label='A')

# Рассчитаем средний чек для группы В

plt.plot(cumulative_revenue_b['date'], 
         cumulative_revenue_b['revenue'] / cumulative_revenue_b['orders'], 
         label='B')

plt.title('График динамики кумулятивного среднего чека по группам А и В')
plt.ylabel("Выручка, млн.")
plt.xticks(rotation=30)
plt.legend()
plt.show()
```

### Относительное изменение кумулятивного чека

## Анализ статистической значимости

### Различия в конверсии между группами, «сырые» данные

Посчитаем статистическую значимость различий в конверсии между группами по «сырым» данным — без удаления аномальных пользователей.

Введем основную и альтернативные гипотезы:

- H0 = различий в конверсии между группами нет
- H1 = различия в конверсии между группами есть
 

Уровень значимости: 0.05

Начнём с подготовки данных.

Так как данные о продажах и средних чеках редко бывают нормально распределены (это классический пример переменных, подверженных выбросам), для проверки гипотез будем использовать непараметрический тест Уилкоксона-Манна-Уитни.

```python
# Выделим пользователей из группы А

visitors_a_daily = visitors[visitors['group'] == 'A'][['date', 'visitors']]
visitors_a_daily.columns = ['date', 'visitors_per_date_a']

visitors_a_cummulative = visitors_a_daily.apply(
    lambda x: visitors_a_daily[visitors_a_daily['date'] <= x['date']].agg(
        {'date': 'max', 'visitors_per_date_a': 'sum'}
    ),
    axis=1,
)

# Соберем пользователей группы А по датам

visitors_a_cummulative.columns = ['date', 'visitors_cummulative_a']
```

```python
# Выделим пользователей из группы В

visitors_b_daily = visitors[visitors['group'] == 'B'][['date', 'visitors']]
visitors_b_daily.columns = ['date', 'visitors_per_date_b']

visitors_b_cummulative = visitors_b_daily.apply(
    lambda x: visitors_b_daily[visitors_b_daily['date'] <= x['date']].agg(
        {'date': 'max', 'visitors_per_date_b': 'sum'}
    ),
    axis=1,
)

# Соберем пользователей группы В по датам

visitors_b_cummulative.columns = ['date', 'visitors_cummulative_b']
```

```python
# Выделим заказы группы А

orders_a_daily = (
    orders[orders['group'] == 'A'][['date', 'transaction_id', 'visitor_id', 'revenue']]
    .groupby('date', as_index=False)
    .agg({'transaction_id': pd.Series.nunique, 'revenue': 'sum'})
)
orders_a_daily.columns = ['date', 'orders_per_date_a', 'revenue_per_date_a']

# Соберем заказы группы А в одну таблицу по дням

orders_a_cummulative = orders_a_daily.apply(
    lambda x: orders_a_daily[orders_a_daily['date'] <= x['date']].agg(
        {'date': 'max', 'orders_per_date_a': 'sum', 'revenue_per_date_a': 'sum'}
    ),
    axis=1,
).sort_values(by=['date'])
orders_a_cummulative.columns = [
    'date',
    'orders_cummulative_a',
    'revenue_cummulative_a',
]
```

```python
# Выделим заказы группы В

orders_b_daily = (
    orders[orders['group'] == 'B'][['date', 'transaction_id', 'visitor_id', 'revenue']]
    .groupby('date', as_index=False)
    .agg({'transaction_id': pd.Series.nunique, 'revenue': 'sum'})
)
orders_b_daily.columns = ['date', 'orders_per_date_b', 'revenue_per_date_b']

# Соберем заказы группы В в одну таблицу по дням

orders_b_cummulative = orders_b_daily.apply(
    lambda x: orders_b_daily[orders_b_daily['date'] <= x['date']].agg(
        {'date': 'max', 'orders_per_date_b': 'sum', 'revenue_per_date_b': 'sum'}
    ),
    axis=1,
).sort_values(by=['date'])

orders_b_cummulative.columns = [
    'date',
    'orders_cummulative_b',
    'revenue_cummulative_b',
]
```

```python
# Теперь соберем поученные данные в общую таблицу

data = (
    orders_a_daily.merge(
        orders_b_daily, left_on='date', right_on='date', how='left'
    )
    .merge(orders_a_cummulative, left_on='date', right_on='date', how='left')
    .merge(orders_b_cummulative, left_on='date', right_on='date', how='left')
    .merge(visitors_a_daily, left_on='date', right_on='date', how='left')
    .merge(visitors_b_daily, left_on='date', right_on='date', how='left')
    .merge(visitors_a_cummulative, left_on='date', right_on='date', how='left')
    .merge(visitors_b_cummulative, left_on='date', right_on='date', how='left')
)

data.head()
```

```python
# Зададим переменные количесвто заказов на пользователя по группам

orders_by_users_a = (
    orders[orders['group'] == 'A']
    .groupby('visitor_id', as_index=False)
    .agg({'transaction_id': pd.Series.nunique})
)
orders_by_users_a.columns = ['visitor_id', 'orders']

orders_by_users_b = (
    orders[orders['group'] == 'B']
    .groupby('visitor_id', as_index=False)
    .agg({'transaction_id': pd.Series.nunique})
)
orders_by_users_b.columns = ['visitor_id', 'orders']
```

Объявим переменные sample_a и sample_b, в которых пользователям из разных групп будет соответствовать количество заказов. Тем, кто ничего не заказал, будут соответствовать нули. Это нужно, чтобы подготовить выборки к проверке критерием Манна-Уитни.

```python
sample_a = pd.concat(
    [orders_by_users_a['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitors_per_date_a'].sum() - len(orders_by_users_a['orders'])
            ),
            name='orders',
        ),],
    axis=0,
)

sample_b = pd.concat(
    [orders_by_users_b['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitors_per_date_b'].sum() - len(orders_by_users_b['orders'])
            ),
            name='orders',
        ),],
    axis=0,
)
```

Напишем функцияю для определения статистической значимости

```python
# Функция для проверки гипотезы о равенстве групп data A и data B
def stat_significance(data_a, data_b):
    alpha = 0.05
    p_value = stats.mannwhitneyu(data_a, data_b)[1]
    print("P-value: {0:.3f}".format(p_value))

    if (p_value < alpha):
        print("Отвергаем нулевую гипотезу: между группами есть разница")
    else:
        print("Не получилось отвергнуть нулевую гипотезу, нет оснований считать группы разными")
    
    print("Относительный прирост В к А: {0:.3%}".format(data_b.mean() / data_a.mean()-1))
```

```python
stat_significance(sample_a, sample_b)
```

### Различия в среднем чеке заказа между группами, «сырые» данные

Посчитаем статистическую значимость различий в среднем чеке между группами по «сырым» данным — без удаления аномальных пользователей.

Введем основную и альтернативные гипотезы:

- H0 различий в среднем чеке между группами нет;
- H1 различия в среднем чеке между группами есть.
 

Уровень значимости: 

Чтобы рассчитать статистическую значимость различий в среднем чеке, передадим критерию mannwhitneyu() данные о выручке с заказов. А ещё найдём относительные различия в среднем чеке между группами.

```python
stat_significance(orders[orders['group']=='A']['revenue'], orders[orders['group']=='B']['revenue'])
```

По неочищенным данным различия в среднем чеке нет.

P-value значительно больше 0.05. Значит, мы также не отвергаем нулевую гипотезу о том, что различий в среднем чеке нет. 

Относительный прирост группы В к группе А составил 28,66%, но оно не является статисстически значимым.

### Различия в конверсии между группами по «очищенным» данным

Так как мы установили наш верхние границы от 4 заказов и 58233. Отсечем по этим значениям данные с выбросами.

```python
# Как мы уже определили выше зададим верхние границы по нашим данным

visitors_with_many_orders = np.percentile(orders_by_users['orders'], 99)
visitors_with_expensive_orders = np.percentile(orders['revenue'], 99)
```

```python
# Отберем данные по количеству заказов

users_with_many_orders = pd.concat(
    [
        orders_by_users_a[orders_by_users_a['orders'] > visitors_with_many_orders]['visitor_id'],
        orders_by_users_b[orders_by_users_b['orders'] > visitors_with_many_orders]['visitor_id'],
    ],
    axis=0,
)

# Отберем данные по сумме заказа

users_with_expensive_orders = orders[orders['revenue'] > visitors_with_expensive_orders]['visitor_id']
abnormal_users = (
    pd.concat([users_with_many_orders, users_with_expensive_orders], axis=0)
    .drop_duplicates()
    .sort_values()
)

# Объеденим в единую таблицу данные по выбросам

abnormal_users.head()
```

```python
#  Посмотрим сколько "ненормальных" пользователей у нас получилось

print(len(abnormal_users))
```

16 - немного, но сейчас посмотрим как сильно эти данные влияли на нашу оценку.

```python
# Зададим переменные с данными очищенными от выбросов по группам

sample_a_filtered = pd.concat(
    [
        orders_by_users_a[
            np.logical_not(orders_by_users_a['visitor_id'].isin(abnormal_users))
        ]['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitors_per_date_a'].sum() - len(orders_by_users_a['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)

sample_b_filtered = pd.concat(
    [
        orders_by_users_b[
            np.logical_not(orders_by_users_b['visitor_id'].isin(abnormal_users))
        ]['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitors_per_date_b'].sum() - len(orders_by_users_a['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)
```

```python
# Посмотрим стат значимость с очищенными данными

stat_significance(sample_a_filtered, sample_b_filtered)
```

По очищенным данным различия в конверсии есть.

P-value меньше 0.05. Значит, мы отвергаем нулевую гипотезу о том, что различий в конверсии нет. 

При этом мы видим значительный относительный конверсии группы В к группе А в 18,401%.

### Различия в среднем чеке заказа между группами, «очищенные» данные

Подсчитаем статистическую значимость различий в среднем чеке после удаления аномальных пользователей.

```python
# Убираем пользователей с завышенными значениями и проверим есть ли различия в среднем чеке между группами

stat_significance(
    orders[(orders['group']=='A') & np.logical_not(orders['visitor_id'].isin(abnormal_users))]['revenue'], 
    orders[(orders['group']=='B') & np.logical_not(orders['visitor_id'].isin(abnormal_users))]['revenue']
                  )
```

По очищенным данным различия в среднем чеке нет.

Мы видим, что группа А показала относительный прирост в 3,234%, однако, мы не можем считать этот результат статистически значимым, потому что значение P-value значительно больше 0,05, хоть и снизилось относительно сырых данных.
